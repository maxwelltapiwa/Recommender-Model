# Building a Recommender System for item purchased in jet using Collaborative Filtering

```python
df.columns
```

```python
features=['Customer ID','Item Purchased','Review Rating']
```

```python
from surprise import Dataset
from surprise import Reader
```

```python
from surprise import Dataset
from surprise import Reader
```

## determining rating scale

```python
df['Review Rating'].max()
```

```python
df['Review Rating'].min()
```

```python
# choosing 2 as lowerbound and 5 as upperbound for rating(ie using integers for bounding)...
reader = Reader(rating_scale=(2, 5))
```

```python
data = Dataset.load_from_df(df[features], reader)
```

## Building model using item-based cosine similarity

```python

from surprise import KNNWithMeans

# here is where item-based is specified
sim_options = {
    "name": "cosine",
    "user_based": False,  # Compute  similarities between items
}
algo = KNNWithMeans(sim_options=sim_options)
```

## getting training set...

```python
trainingSet = data.build_full_trainset()
```

```python
algo.fit(trainingSet)
```

## testing model...

```python
prediction1 = algo.predict(1,'Blouse')
```

```python
prediction1.est
```

```python
def estimate(entry):
    prediction = algo.predict(*entry)
    return prediction.est
```

```python
estimate((2,'Sweater'))
```

```python
estimate((4,'Sandals'))
```

# Results:

## out of the three tested, all products had correct predictions with:

- 1,'Blouse' $\rightarrow$ 3.1 (correct)
- 2,'Sweater' $\rightarrow$ 3.1 (correct) and
- 4,'Sandals' $\rightarrow$ 3.1 (correct)

futher testing the model will show that the predictions will not always be on point with the actual ratings but will still be close, eg 3.1 for an actual rating of 3.5
